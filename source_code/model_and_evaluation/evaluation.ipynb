```python
def find_conn_string(input_str, target):
    input_list = input_str.split()
    target_list = target.split()
    
    input_index = len(input_list) - 1
    target_index = len(target_list) - 1
    
    while(input_index > -1 and target_index > -1):
        if input_list[input_index] == target_list[target_index]:
            target_index -= 1
            input_index -= 1
        else:
            target_list = target_list[:target_index+1]
            input_list = input_list[:input_index+1]
            break
            
#     print("state 1: target_list: {}".format(target_list))
#     print("state 2: input_list: {}".format(input_list))
    
    while((target_index-2) > -1):
        if target_list[target_index] in input_list and \
            target_list[target_index - 1] in input_list and \
            target_list[target_index - 2] in input_list:
            break
        else:
            target_index -= 1
    
    if (target_index + 1) < len(target_list):
        part_of_list = target_list[target_index+1:]
    else:
        part_of_list = []
    
    target_list = target_list[:target_index+1]
    updated_target_list = []
    
    for word in target_list:
        if word in input_list:
            input_list.remove(word)
        else:
            updated_target_list.append(word)
            
    return " ".join(updated_target_list + part_of_list)


input_ = "it has exactly what one needs and the quality is perfect too. the bag is little small."
target_ = "it has exactly what one needs and the quality is perfect too. on, the bag is little small."

print(find_conn_string(input_[:], target_[:]))
```
