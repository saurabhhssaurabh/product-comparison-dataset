```python
verb_list = [AUX, VERB]
noun_list = [NOUN, PRON, PROPN]
first_word_pos = [NOUN, PRON, PROPN, DET, ADP, ADJ]

first_token_list = ["because", "and", "before", "but", "however", "now", "of", "then", "&", "or", "that", "although"]
first_person_and_pronoun_words = ["i", "me", "my", "myself", "mine", "we", "us", "our", "ourselves", \
                                  "he", "she", "him", "her", "am", "they", "them", "themselves", "their", "theirs", "you",\
                                 "yours", "your"]
first_word_skip_segment = ["who", "what", "when", "whom", "how", "whenever", "where", "which", "whose", "why", "after",\
                           "hey", "to", "in", "if", "so"]
```

```python
def pre_process_and_validation_check(segment):
    
    token_list = list(en_spacy(segment))
    
    # remove leading and trailing punctuations.
    # if first token is in `first_token_list`, remove it.
    while(len(token_list) > 0 and ((token_list[0].text in string.punctuation) or \
                                   (token_list[0].text in first_token_list))):
        if len(token_list) == 1:
            token_list = []
        else:
            token_list = token_list[1:]   
        
    while(len(token_list) > 0 and token_list[-1].text in string.punctuation):
        if len(token_list) == 1:
            token_list = []
        else:
            token_list = token_list[:-1]
    
    # if number of tokens are less than 2, skip the segment
    if len(token_list) <= 2:
        return None, None
    
    # if first word is a verb, skip the segment
    if token_list[0].pos in verb_list:
        return None, None

    # if first word in first_word_skip_segment list, skip the segment
    if token_list[0].text in first_word_skip_segment:
        return None, None

    # If first person word in a segment, skip the segment
    is_found = False
    for token in token_list:
        if token.text in first_person_and_pronoun_words:
            is_found = True
            break
            
    if is_found:
        return None, None
    
    # Check whether a segment has at least one noun and one verb
    is_noun = False
    is_verb = False

    tmp_list = []
    
    for token in token_list:
        tmp_list.append(token.pos)
        if token.pos in noun_list:
            is_noun = True
        elif token.pos in verb_list:
            is_verb = True

    if (not is_noun) or (not is_verb):
        return None, None
    
    return token_list, tuple(tmp_list)

sent = "to recharge it".strip().lower()
test_token_list, test_pos_tuple = pre_process_and_validation_check(sent)
print(test_token_list)
print(test_pos_tuple)
```

```python
def is_considerable(spacy_token_list, debug=False):
    noun_detected = False
    valid_seg = True
    
    if debug:
        for token in spacy_token_list:
            print("{} : {}".format(token.text, token.pos_))

    if spacy_token_list[0].pos in first_word_pos and \
        spacy_token_list[-1].pos not in verb_list:
            for token in spacy_token_list:
                if token.pos in noun_list:
                    noun_detected = True
                elif (token.pos in verb_list) and (not noun_detected):
                    valid_seg = False
                    break
    else:
        valid_seg = False
                    
    return valid_seg

# print(is_considerable(test_token_list, True))
```

```python
import json, os

pattern_dict = {}
pattern_dict[0] = "{} . however , there are people who have complained that {} ."
pattern_dict[1] = "{} . but , {} ."
pattern_dict[2] = "{} . on the other hand , {} ."
pattern_dict[3] = "{} . however {} ."
pattern_dict[4] = "{} . yet , some users have also mentioned that {} ."
pattern_dict[5] = "although {} , according to a few users {} ."
pattern_dict[6] = "{} . on the other hand , a few users have complained that {} ."

ignore_file_list = ["0684873176"]


def make_references(positive_op, negative_op):
    ref_list = []

    for key in pattern_dict.keys():
        ref_list.append(pattern_dict.get(key).format(positive_op, negative_op))

    return ref_list


def generate_dataset(pos_dir, neg_dir, train_file, val_file, test_file):
    counter = 1
    file_counter = 0
    
    train_f = open(train_file, "w")
    val_f = open(val_file, "w")
    test_f = open(test_file, "w")

    for file in os.listdir(pos_dir):
        
        if file in ignore_file_list:
            continue
        
        file_counter += 1
        pos_op_list = []
        with open(os.path.join(pos_dir, file), "r") as in_f:
            for line in in_f:
                pos_op_list.append(line.strip().lower())

        neg_op_list = []
        if os.path.exists(os.path.join(neg_dir, file)):
            with open(os.path.join(neg_dir, file), "r") as in_f:
                for line in in_f:
                    neg_op_list.append(line.strip().lower())
        else:
            continue

        for pos_op in pos_op_list:
            for neg_op in neg_op_list:
                tmp_dict = {}
                tmp_dict["input"] = pos_op + " . " + neg_op + " ."
                tmp_dict["gt_output"] = pattern_dict[counter%7].format(pos_op, neg_op)
                tmp_dict["connecting_string"] = pattern_dict[counter%7].format("", "").strip(". ")
                tmp_dict["gt_references"] = make_references(pos_op, neg_op)
                tmp_dict["product_id"] = file

                counter += 1
                
                if file_counter <= 2569:
                    train_f.write(json.dumps(tmp_dict)+"\n")
                elif file_counter <= 2890:
                    val_f.write(json.dumps(tmp_dict)+"\n")
                else:
                    test_f.write(json.dumps(tmp_dict)+"\n")
                    
    train_f.close()
    val_f.close()
    test_f.close()
                    
pos_dir = "/data3/saurabh/Product-Comparison/data/v6_dataset/5_qt_data/pos_summary_last_word_filter/amazon_electronics_3269_run_1"
neg_dir = "/data3/saurabh/Product-Comparison/data/v6_dataset/5_qt_data/neg_summary_last_word_filter/amazon_electronics_3269_run_1"
train_file = "/data3/saurabh/Product-Comparison/data/v6_dataset/6_dataset/amazon_electronics_3269_run_1/train.jsonl"
val_file = "/data3/saurabh/Product-Comparison/data/v6_dataset/6_dataset/amazon_electronics_3269_run_1/val.jsonl"
test_file = "/data3/saurabh/Product-Comparison/data/v6_dataset/6_dataset/amazon_electronics_3269_run_1/test.jsonl"

generate_dataset(pos_dir, neg_dir, train_file, val_file, test_file)
```

```python
def read_jsonl(file_name):
        input_data = []
        output_data = []

        with open(file_name, "r") as in_f:
            for line in in_f:
                tmp_dict = json.loads(line)
                input_data.append(tmp_dict["input"])
                output_data.append(tmp_dict["gt_output"])

        return input_data, output_data



class ComparativeResponseDataset(torch.utils.data.Dataset):
    
    def __init__(self, input_encoding, output_encoding):
        self.input_encoding = input_encoding
        self.output_encoding = output_encoding
        
    def __getitem__(self, index):
        item = {}
        item["input_ids"] = np.array(self.input_encoding.input_ids[index])
        item["attention_mask"] = np.array(self.input_encoding.attention_mask[index])
        item["decoder_input_ids"] = np.array(self.output_encoding.input_ids[index])
        item["decoder_attention_mask"] = np.array(self.output_encoding.attention_mask[index])
        item["labels"] = np.array(self.output_encoding.input_ids[index])
                
        return item
    
    def __len__(self):
        return len(self.input_encoding.input_ids)
```
